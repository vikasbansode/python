### Part 3 : Axes,labels,legends,Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days  = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
```

### Part 4 : Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGLE','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

### Part 5 : Histograms 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g')
```

```python
bld_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
bld_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([bld_sugar_men,bld_sugar_women],
         bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],
        label=['men','women'],orientation='horizontal')
plt.legend()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals =[1400,600,300,410,250]
exp_labes = ["Home Rent","Food","Phone/Internet Bill","Car","Other Untilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labes,radius = 1.5,autopct='%0.2f%%',
       shadow=True,explode=[0,0.1,0.1,0,0],startangle=180)
plt.show()
```

### Part 7 : Save Chat to a file using savefig

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labes,radius = 1.5,autopct='%0.2f%%',
       shadow=True,explode=[0,0.1,0.1,0,0],startangle=180)
plt.savefig('piechart.png',bbox_inches='tight',pad_inches=2,transparent=True)
```
