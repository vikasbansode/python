```python
import pandas as pd
def convert_people_cell(cell):
    if cell=='n.a':
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell == 'Not available':
        return None
    return cell

df =pd.read_excel("stock_rates.xlsx","stock_rates",converters={
    'people' : convert_people_cell,
    'eps' : convert_eps_cell
})
df
```
